```python
bat = pd.read_csv("../data/raw/mlb/war_daily_bat.txt")
pit = pd.read_csv("../data/raw/mlb/war_daily_pitch.txt")

people = pd.read_csv("../data/raw/mlb/People.csv")

def find_col(df, candidates):
    """Return the first matching column from candidates (case-insensitive, exact or contains)."""
    cols = df.columns.tolist()
    low = {c: c.lower() for c in cols}
    for cand in candidates:
        cand_low = cand.lower()
        # exact match
        for c in cols:
            if low[c] == cand_low:
                return c
        # contains match
        for c in cols:
            if cand_low in low[c]:
                return c
    return None

def standardize_war_table(df, kind):
    """
    Standardize BR WAR table to columns:
      bbrefID, season, <kind>_bwar
    """
    id_col = find_col(df, ["player_id", "playerid", "bbrefid", "key_bbref", "bbref_id", "id"])
    year_col = find_col(df, ["year_id", "yearid", "season", "year"])
    war_col = find_col(df, ["war", "WAR"])

    if id_col is None or year_col is None or war_col is None:
        raise ValueError(
            f"Could not detect required columns in {kind} table.\n"
            f"Detected -> id_col={id_col}, year_col={year_col}, war_col={war_col}\n"
            f"Columns available: {df.columns.tolist()}"
        )

    out = df[[id_col, year_col, war_col]].copy()
    out = out.rename(columns={id_col: "bbrefID", year_col: "season", war_col: f"{kind}_bwar"})

    # Clean types
    out["bbrefID"] = out["bbrefID"].astype(str).str.strip()
    out["season"] = pd.to_numeric(out["season"], errors="coerce").astype("Int64")
    out[f"{kind}_bwar"] = pd.to_numeric(out[f"{kind}_bwar"], errors="coerce")

    out = out.dropna(subset=["bbrefID", "season"])
    out["season"] = out["season"].astype(int)

    return out

bat_m = standardize_war_table(bat, "bat")
pit_m = standardize_war_table(pit, "pit")

print("Standardized batting WAR sample:")
display(bat_m.head())
print("Standardized pitching WAR sample:")
display(pit_m.head())

# Map bbrefID -> Lahman playerID
# Lahman People.csv usually includes a 'bbrefID' column
people_cols = [c.lower() for c in people.columns]
if "bbrefid" not in people_cols:
    raise ValueError(f"People.csv missing bbrefID column. Columns: {people.columns.tolist()}")

# Get actual casing
people_bbref_col = [c for c in people.columns if c.lower() == "bbrefid"][0]
people_pid_col   = [c for c in people.columns if c.lower() == "playerid"][0]

id_map = (people[[people_pid_col, people_bbref_col]]
          .dropna()
          .rename(columns={people_pid_col: "player_id", people_bbref_col: "bbrefID"}))

id_map["bbrefID"] = id_map["bbrefID"].astype(str).str.strip()
id_map["player_id"] = id_map["player_id"].astype(str).str.strip()

bat_m = bat_m.merge(id_map, on="bbrefID", how="left").dropna(subset=["player_id"])
pit_m = pit_m.merge(id_map, on="bbrefID", how="left").dropna(subset=["player_id"])

# Combine into season-level WAR
bwar = pd.merge(
    bat_m[["player_id", "season", "bat_bwar"]],
    pit_m[["player_id", "season", "pit_bwar"]],
    on=["player_id", "season"],
    how="outer"
)

bwar[["bat_bwar", "pit_bwar"]] = bwar[["bat_bwar", "pit_bwar"]].fillna(0.0)
bwar["bwar"] = bwar["bat_bwar"] + bwar["pit_bwar"]

print("Season-level bWAR sample:")
display(bwar.head())

# Merge with utility score table
utility = pd.read_csv("../data/processed/season_utility_scores.csv")
utility["player_id"] = utility["player_id"].astype(str).str.strip()
utility["season"] = utility["season"].astype(int)

season_df = utility.merge(
    bwar[["player_id", "season", "bwar", "bat_bwar", "pit_bwar"]],
    on=["player_id", "season"],
    how="left"
)

season_df[["bwar", "bat_bwar", "pit_bwar"]] = season_df[["bwar", "bat_bwar", "pit_bwar"]].fillna(0.0)

print("Merged season_df sample:")
display(season_df.head())

# Sanity checks
print("bWAR summary:")
print(season_df["bwar"].describe())

missing_rate = (season_df["bwar"].isna().mean())
print("Missing bWAR rate (should be 0 after fill):", missing_rate)
```
